```python
#q4:ans
#exception LookupError
#This is the base class for those exceptions that are raised when a key or index used on a mapping or sequence is 
#invalid or not found. The exceptions raised are :

#IndexError
#the IndexError is a common exception that occurs when trying to access an element in a list, tuple, or any other sequence using an index that is outside the valid range of indices for that sequence..
#example
try:
    a=[1,2,3]
    print(a[4])
except LookupError:
    print('index error occure')
else:
    print('success')
    
#key error
#A Python KeyError exception is what is raised when you try to access a key that isn’t in a dictionary (dict).
try:
    a={'a1':1,'b1':2}
    print(a['a3'])
except LookupError:
    print('key error occur')
else:
    print('success')
```
